```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
# drive.mount('/content/drive')

arr  = pd.read_csv('/content/drive/My Drive/GC/headbrain.csv')

print(arr.shape)
print(arr.head)
from sklearn.model_selection import train_test_split
X = arr['Head Size(cm^3)'].values
Y = arr['Brain Weight(grams)'].values

X_train , X_test , Y_train , Y_test = train_test_split(X , Y ,test_size = 0.3 , random_state = 2)

X_mean = np.mean(X_train)
Y_mean = np.mean(Y_train)
n = len(X_train)

num = 0
den = 0
for i in range(n):
    num += (X_train[i]-X_mean) * (Y_train[i] - Y_mean)
    den += (X_train[i]-X_mean) ** 2

B1 = num/den
B0 = Y_mean - B1 * X_mean
print("B0 = ",B0)
print("B1 = ",B1)

x_min = np.min(X_train)
x_max = np.max(X_train)

x = np.linspace(x_min , x_max , num = 100)
y = B1 * x + B0

plt.scatter(X_train,Y_train ,color = 'r' ,label = "Data Point")
plt.plot(x,y ,color = 'b' , label = "Linear reg")
plt.xlabel("Head Size(cm^3)")
plt.ylabel("Brain Weight(grams)")
plt.legend()
plt.show()



```

```python
x_min = np.min(X_test)
x_max = np.max(X_test)

x = np.linspace(x_min , x_max , num = 100)
y = B1 * x + B0

plt.scatter(X_test,Y_test ,color = 'r' ,label = "Data Point")
plt.plot(x,y ,color = 'b' , label = "Linear reg")
plt.xlabel("Head Size(cm^3)")
plt.ylabel("Brain Weight(grams)")
plt.legend()
plt.show()
```

```python
# RSS and R2 for training data
def Y_pred(x):
    return B1 * x + B0
n = len(X_train)
rss = 0
for i in range(n):
    rss += (Y_train[i] - Y_pred(X_train[i])) **2
rmse = np.sqrt(rss/n)
Y_mean = np.mean(Y_train)
tss = sum((Y_train - np.ones(n)*Y_mean)**2)
R2 = 1 -(rss/tss)
print("RSS = ",rss)
print("R2 = ",R2)
```
